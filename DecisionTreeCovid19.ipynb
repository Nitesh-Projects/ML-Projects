```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/python/Covid_Latest/allcovid19India.csv")
```

```python
data = data.iloc[:,1:]
```

```python
data
```

```python
day =data[data['Current Status']=='Hospitalized'].groupby(['Month','Day'])[['Num Cases']].sum()
```

```python
day
```

```python
x = np.arange(len(day))
```

```python
x = x.reshape(-1,1)
x
```

```python
y = day.values
```

```python
y
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg = DecisionTreeRegressor()
```

```python
reg.fit(x,y)
```

```python
Yp = reg.predict(x)
```

```python
Yp
```

```python
plt.scatter(x,y)
plt.plot(x,Yp,'k',10)
```

```python
reg.score(x,y)*100
```
