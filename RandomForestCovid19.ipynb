```python
day =data[data['Current Status']=='Hospitalized'].groupby(['Month','Day'])[['Num Cases']].sum()
```

```python
x= np.arange(len(day))
```

```python
x = x.reshape(-1,1)
```

```python
y = day.values
```

```python
y
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg = RandomForestRegressor( n_estimators=30)
```

```python
reg.fit(x,y)
```

```python
#n=50
reg.score(x,y)*100
```
